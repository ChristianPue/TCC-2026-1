```python
from ortools.sat.python import cp_model
```

```python
model = cp_model.CpModel()
```

```python
# variables
x1 = model.NewIntVar(1, 2, 'x1')
x2 = model.NewIntVar(1, 2, 'x2')
x3 = model.NewIntVar(1, 3, 'x3')
```

```python
# restricciones:
model.Add(3 * x1 + x2 + x3 == 10)
```

```python
model.AddAllDifferent([x1, x2, x3]) # todos los valores deben ser distintos
```

```python
solver = cp_model.CpSolver()
status = solver.Solve(model)
```

```python
if status == cp_model.FEASIBLE or status == cp_model.OPTIMAL:
  print('Solución encontrada')
  print(f'x1: {solver.value(x1)}, \nx2: {solver.value(x2)}, \nx3: {solver.value(x3)}')
else:
  print('No hay solución')
```
