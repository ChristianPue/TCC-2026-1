## Probar con with lines

## Con datos en más de 3 columnas

````python
!wget https://litis.univ-lehavre.fr/~guinand/Guides/data/data3
````

````python
gnuplot_script3 = """
set terminal png size 900,600
set output 'plot_multi.png'

set title 'Datos con más de 3 columnas'
set xlabel 'X (columna 1)'
set ylabel 'Valores'

plot "data3.dat" using 1:2 with lines title 'Columna 2', \
     "data3.dat" using 1:3 with lines title 'Columna 3', \
     "data3.dat" using 1:4 with lines title 'Columna 4'

"""

with open('script_multi.plt', 'w') as f:
    f.write(gnuplot_script3)
````

````python
subprocess.run(['gnuplot', 'script_multi.plt'])
Image('plot_multi.png')
````

## Antes de plot colocar: set xrange[100:200] ; set yrange[0:900]



```
set xrange[100:200] ; set yrange[0:900]
```



````python
gnuplot_script4 = """
set terminal png size 900,600
set output 'plot_rango.png'

set title 'Gráfico con rango personalizado'
set xlabel 'X'
set ylabel 'Y'

set xrange [100:200]
set yrange [0:900]

plot "data3.dat" using 1:2 with lines title 'Col 2', \
     "data3.dat" using 1:3 with lines title 'Col 3', \
     "data3.dat" using 1:4 with lines title 'Col 4'

"""

with open('script_rango.plt', 'w') as f:
    f.write(gnuplot_script4)
````

````python
subprocess.run(['gnuplot', 'script_rango.plt'])
Image('plot_rango.png')
````
