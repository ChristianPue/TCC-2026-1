````python
import subprocess
from IPython.display import Image

gnuplot_script = """
set terminal png
set output 'plot.png'
plot sin(x)
"""
with open('script.plt', 'w') as f:
    f.write(gnuplot_script)
````

o también




## Con títulos





```
plot [-pi:pi] sin(x) with impulses notitle, \
     sin(x) with lines ls 1 title "sinus", \
     cos(x) w filledcurves fs transparent pattern 4 title "cosinus"
```



````python
gnuplot_script = """
set terminal png
set output 'plot10.png'
plot [-pi:pi] sin(x) with impulses notitle, \
     sin(x) with lines ls 1 title "seno(X)", \
     cos(x) w filledcurves fs transparent pattern 4 title "coseno(X)"
"""
with open('script10.plt', 'w') as f:
    f.write(gnuplot_script)
````

````python
subprocess.run(['gnuplot', 'script10.plt'])
Image('plot10.png')
````
