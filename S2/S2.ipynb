## Probar con with lines

````python
gnuplot_script = """
set terminal png size 800,600
set output 'plot12.png'
set title 'Plot con lines'
set xlabel 'X'
set ylabel 'Y'
plot "demo-gnuplot.data" using 1:2 with lines title 'Con Datos'
"""

with open('script12.plt', 'w') as f:
  f.write(gnuplot_script)

subprocess.run(['gnuplot', 'script12.plt'])
Image('plot12.png')
````

````python
gnuplot_script_steps = """
set terminal png size 800,600
set output 'plot13.png'
set title 'Plot con steps (escalones)'
set xlabel 'X'
set ylabel 'Y'
plot "demo-gnuplot.data" using 1:2 with steps title 'Con Datos (steps)'
"""

with open('script13.plt', 'w') as f:
  f.write(gnuplot_script_steps)

subprocess.run(['gnuplot', 'script13.plt'])
Image('plot13.png')
````

## Con datos en más de 3 columnas

````python
!wget https://litis.univ-lehavre.fr/~guinand/Guides/data/data3
````

````python
gnuplot_script = """
set terminal pngcairo size 1000,700 enhanced
set output 'plot14.png'

set title 'Datos con más de 3 columnas'
set grid
set xlabel 'Columna 1'
set ylabel 'Valores'

plot 'data3' using 1:2 with lines title 'Columna 2', \
  'data3' using 1:3 with lines title 'Columna 3', \
  'data3' using 1:4 with lines title 'Columna 4'

set output
"""

with open('script14.plt', 'w') as f:
  f.write(gnuplot_script)

subprocess.run(['gnuplot', 'script14.plt'])

display(Image('plot14.png'))
````

## Antes de plot colocar: set xrange[100:200] ; set yrange[0:900]



```
set xrange[100:200] ; set yrange[0:900]
```



````python
gnuplot_script = """
set terminal pngcairo size 1000,700 enhanced
set output 'plot15.png'

# === Configuración de rangos ===
set xrange[100:200]
set yrange[0:900]

set title 'Datos con rangos personalizados (xrange y yrange)'
set grid
set xlabel 'Columna 1'
set ylabel 'Valores'

plot 'data3' using 1:2 with lines title 'Columna 2', \
  'data3' using 1:3 with lines title 'Columna 3', \
  'data3' using 1:4 with lines title 'Columna 4'

set output
"""

with open('script15.plt', 'w') as f:
  f.write(gnuplot_script)

subprocess.run(['gnuplot', 'script15.plt'])

display(Image('plot15.png'))
````

## Probar colocando títulos



```
plot ’demo-gnuplot.data’ with lines
set grid
set xlabel "molecular weight"; set ylabel "volumes"
```



````python
gnuplot_script = """
set terminal pngcairo size 900,650 enhanced
set output 'plot16.png'

# Configuraciones solicitadas
set grid
set xlabel "molecular weight"
set ylabel "volumes"

# Título del gráfico
set title "Gráfico con títulos y etiquetas"

# El plot
plot 'demo-gnuplot.data' with lines title 'Datos experimentales'

set output
"""

with open('script16.plt', 'w') as f:
  f.write(gnuplot_script)

subprocess.run(['gnuplot', 'script16.plt'])

display(Image('plot16.png'))
````
