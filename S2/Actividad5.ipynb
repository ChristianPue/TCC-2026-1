```python
!pip install dbfread -q

from dbfread import DBF
import pandas as pd

# Leer el archivo dbf
db = DBF('IE_P.dbf', encoding='latin1')   # o 'utf-8' si da error

# Mostrar columnas
print("=== COLUMNAS DEL ARCHIVO ===")
print([field.name for field in db.fields])

# Mostrar las primeras 5 filas como tabla
df = pd.DataFrame(iter(db))
print("\n=== PRIMERAS 5 FILAS ===")
print(df.head().to_string())

# Guardar todo en CSV para usar después en GNUPlot
df.to_csv('ie_p_data.csv', index=False, encoding='utf-8')
print("\n✅ Archivo CSV guardado como 'ie_p_data.csv'")
```
